# 1-1　期待値

 ### 期待値の定義

期待値とは確率変数の試行を十分に行い、その結果として得られる数値の平均の事である. すなわち、試行によって得られる数値Xを$x_{1}$,$x_{2}$,$x_{3}$, …, $x_{n}$、それぞれの値の確率を$p_{1}$,$p_{2}$,$p_{3}$, … ,$p_{n}$とすると、Xの期待値E(X)は<br> $$　E(X) = \sum^{n}_{k=1} x_{n}p_{n} $$となる.

このページでは期待値についての説明をする. 例としてサイコロを使って話を行う. サイコロを振って出た目の平均値と、定義から求めた期待値とを比較することでこの定義の確認を行う.


***


### ① サイコロを10回振るときの平均値

初めにサイコロを10回振るときの平均値を求めてみる. まずサイコロの出目の再現のため、サイコロと同じ確率変数のコードを作成した. このコードでは1~6までの値がそれぞれ1/6の確率で出る. つまり全ての目が等確率で出るコードである. このサイコロのコードを10回試行し、その結果から平均値を求めてみる. またそれぞれの出た目の回数を視覚化するため棒グラフと円グラフを作成した. 

```python
# matplotlib をインラインで使用することを指定する
%matplotlib inline
# ライブラリの読み込み
import matplotlib.pyplot as plt
import random
#出る目を固定
random.seed(2)
#サイコロを振る回数、今回は10回
try_num = 10
#サイコロの目は1-6の範囲であるから1-6の間で数値が与えられるように
appearances = []
for i in range(try_num):
    appearances.append(random.randint(1,6))
```

```python
#結果の棒グラフ、一目見た時にどの目がいくら出たかが分かりやすい
plt.figure()
plt.subplot(121)
plt.hist(appearances, bins=20)

#結果の円グラフ、こちらはそれぞれの目の出る割合が分かりやすい
rate = []
for i in range(1,7):
    rate.append(appearances.count(i) / try_num)

#円グラフの設定
label = ["1","2","3","4","5","6"]
plt.subplot(122)
plt.pie(rate,labels=label,autopct="%.1f%%",startangle=90,counterclock=False)
plt.show()
```

```python
#平均値を求める
average = sum(appearances) / len(appearances)
print("平均値 =",average)
```

### ② 100万回サイコロを振るときの平均値

次にサイコロのコードを100万回試行したときの平均値を求めてみる. 試行回数を10回から100万回と増加させたことで、出目の平均値や出目の偏りが変化したかを確認する. また変化したと考えられる場合、平均値や出目の偏りがどのように変化したかを確認する. 

```python
#サイコロを振る回数、今回は100万回
try_num = 1000000
for i in range(try_num):
    appearances.append(random.randint(1,6))

#棒グラフ
plt.figure()
plt.subplot(121)
plt.hist(appearances, bins=100)

#円グラフ
rate = []
for i in range(1,7):
    rate.append(appearances.count(i) / try_num)

#円グラフの設定
label = ["1","2","3","4","5","6"]
plt.subplot(122)
plt.pie(rate,labels=label,autopct="%.1f%%",startangle=90,counterclock=False)
plt.show()
```

```python
average = sum(appearances) / len(appearances)
print("平均値 =",'{:.3g}'.format(average))
```

## 結論

全ての目が等確率で出る、理想的なサイコロの目の期待値を計算してみると
\begin{eqnarray}
\ E(X)　= \sum^{n}_{k=1} x_{n}p_{n} &=& 1×\frac{1}{6} + 2×\frac{1}{6} + 3×\frac{1}{6} + 4×\frac{1}{6} + 5×\frac{1}{6} + 6×\frac{1}{6} \\
&=& 3.5
\end{eqnarray}
よって理想的なサイコロの目の期待値は3.5である. ①のグラフを見てみると①の出目には大きな偏りがある. また、①の平均値と理想的なサイコロの期待値は一致していない. 一方、②の出目はほとんど等確率である. さらに、②の平均値は理想的なサイコロの期待値と同じ値になっている. つまりこの定義に沿って考えてみると、ある確率変数の平均値から期待値を求めるには十分な試行が必要であるといえる.
