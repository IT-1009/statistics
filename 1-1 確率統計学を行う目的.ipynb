# 1-1　確率統計学を学ぶ目的とは

統計学は未知の物体や現象に対して、経験的に得られたデータから正確な推測を行うことを目的とする学問である。

この章では統計学の目的を理解するために、ある事象の統計から母集団の性質の推測を行う。

ここではある事象を「サイコロを振ること」とする。サイコロを未知の物質として、サイコロの性質の推測を行う。サイコロは6面で1~6の目がそれぞれ1つの面に書いてあるものを使う。

また、試行回数が統計による推測に及ぼす影響を調べるため、試行回数が十分である時の推測と試行回数が不十分な時の推測を比較する。不十分な回数は10回、十分な回数は1000000回として試行を行う。

### （Ⅰ）① 10回サイコロを振るときの期待値と結果を調べる

まずは試行回数を10回にしてサイコロに対する推測を行う。試行回数が不十分であると推測はどのようになるか。

```python
# matplotlib をインラインで使用することを指定する
%matplotlib inline
# ライブラリの読み込み
import matplotlib.pyplot as plt
import random
#出る目を固定
random.seed(1)
```

```python
#サイコロを振る回数、今回は10回
appearances = []
try_num = 10
```

```python
#サイコロの目は1-6の範囲であるから1-6の間で数値が与えられるように
for i in range(try_num):
    appearances.append(random.randint(1,6))
```

```python
#平均値を求める
average = sum(appearances) / len(appearances)
print(average)
```

10回サイコロを振った時に出る目の期待値が3.4になる事が分かった。
またサイコロの性質の推測を行うために結果をグラフで可視化してみる。

```python
#結果の棒グラフ、一目見た時にどの目がいくら出たかが分かりやすい
plt.hist(appearances, bins=20)
plt.show()
```

```python
#結果の円グラフ、こちらはそれぞれの目の出る割合が分かりやすい
rate = []
for i in range(1,7):
    rate.append(appearances.count(i) / try_num)
#円グラフの設定
label = ["1","2","3","4","5","6"]
plt.subplot(121)
plt.pie(rate,labels=label,autopct="%.1f%%",startangle=90,counterclock=False)
plt.show()
```

### ② 試行回数10回の時の結果

上の2つのグラフより、サイコロは4が多く出るものである。またサイコロを振った時に出る目の期待値は3.4であることが分かったと言える。しかしこれは正しくサイコロの性質をとらえられているのだろうか？

試行回数を100万回に変化させもう一度統計を行ってみる。


### ③ 100万回サイコロを振るときの期待値と結果を調べる

100万回サイコロを振り、その結果をグラフ化する。試行の方法は先ほどと同様にする。

```python
try_num = 1000000 
for i in range(try_num):
    appearances.append(random.randint(1,6))
average = sum(appearances) / len(appearances)
print(average)

plt.figure()
plt.subplot(121)
plt.hist(appearances, bins=100)

rate = []
for i in range(1,7):
    rate.append(appearances.count(i) / try_num)
label = ["1","2","3","4","5","6"]
plt.subplot(122)
plt.pie(rate,labels=label,autopct="%.2f%%",startangle=90,counterclock=False)
plt.show()
```

### ④ 試行回数100万回の時の結果

期待値はほぼ3.5になった。また1~6目の出る確率もそれぞれ等しくなっていると言える。

この結果からサイコロという物体は1~6の目が同じ確率で出る、という特性を持っていることが分かったと言える。

## （Ⅱ）結論

どちらの推測が正確であったか。

サイコロを転がしたとき、理想的なサイコロならば1~6の面がそれぞれ等確率で出るという性質を持つ。そのため、サイコロが出るの目の期待値は3.5になるという事が一般的に知られている。

つまり試行回数10回の時に比べ試行回数100万回の統計を行って得られた推測がより正確である事が分かる。

よって、帰納的な推測においては、ある程度の数のデータを得る事が正しい推測を行うために必要である。

## （Ⅲ）まとめ
これが簡単な推計統計学のイメージである。記述統計を行う事で、ある程度の数があり、かつ規則性のないデータから未知の物体や現象の特性を推測できる。これが確率統計学を行う目的である。

統計学では、特徴を求めたい大きなデータの集合全体の事を「母集団」、母集団の情報を推測するために選ばれた一部の集団の事を「標本」という。

また、標本の特徴から母集団の推測を行う事を推計統計学といい、標本をデータとして文字や表、グラフにして表す事を記述統計学という。
