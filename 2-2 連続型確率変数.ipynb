### 1次元の連続型確率変数とは

### 連続型確率変数
連続型確率変数とはとりうる値が連続的な確率変数のことである. 例えば、「人の身長」のようにとりうる値が連続している確率変数がこれにあたる.

### 確率密度関数
連続型確率変数を表現する際には、とりうる値の区間を[a,b]と定義する必要がある. この時この関数を確率密度関数、あるいは単に密度関数$f(x)$とよぶ. こうした表現をする理由は、離散型確率変数と異なり、確率変数がある値をとる時の確率という定義をしていないからである. そのため、連続型確率変数では$f(x)=P(X=x_{i})$とならないことに注意する必要がある.

***

まず、連続型確率変数の例として、いかさまルーレットを考える. ルーレットは円周の長さを1とし、ルーレットが止まった位置での弧の長さを実現値とする. そのため、このルーレットの取りうる値は0~1の実数となる. また、このルーレットでは大きい数ほど出やすいいかさまが仕組まれているとする. この時密度関数は次のような関数を定義する.
\begin{equation}
f(x)＝
\begin{cases}
  3x & (0 \leqq x \leqq 1)\\\
  0 & (その他)
\end{cases}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

%precision 3
%matplotlib inline
```

```python
x_range = np.array([0, 1])
def f(x):
    if x_range[0] <= x <= x_range[1]:
        return 3 * x
    else:
        return 0
#確率変数Xの定義
X = [x_range, f]
```

```python
xs = np.linspace(x_range[0], x_range[1], 100)

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

ax.plot(xs, [f(x) for x in xs], label='f(x)', color='gray')
ax.hlines(0, -0.2, 2.2, alpha=0.3)
#左の青線
ax.vlines(0, -0.2, 3.2, alpha=0.3)
#右の点線
ax.vlines(xs.max(), 0, 3.2, linestyles=':', color='gray')

# 0.4から0.6のx座標を用意
xs = np.linspace(0.4, 0.6, 100)
# xsの範囲でf(x)とx軸に囲まれた領域を塗りつぶす
ax.fill_between(xs, [f(x) for x in xs], label='確率')

ax.set_xticks(np.arange(-0.2, 1.3, 0.1))
ax.set_xlim(-0.1, 1.1)
ax.set_ylim(-0.2, 3.1)
ax.legend(prop = {"family" : "MS Gothic"})

plt.show()
```
